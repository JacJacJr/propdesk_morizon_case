### Import required packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Read prepared data

```python
housing_data_path = "all_platforms_data.csv"
df_housing = pd.read_csv(filepath_or_buffer=housing_data_path, delimiter=",", header=0, encoding="utf-8")

df_housing.sample(2)
```

### Features analysis

##### Correlation Heatmap

```python
# Convert categorical data to numerical data
import numpy as np
df_housing_numeric = df_housing.select_dtypes(include=[np.number])

correlations = df_housing_numeric.corr()

plt.figure(figsize=(20, 20))
sns.heatmap(correlations, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.3, linecolor='black', annot_kws={"size": 8})
plt.title('Correlation Heatmap', fontsize=12)
plt.xticks(fontsize=8, rotation=45)
plt.yticks(fontsize=8)
plt.show()
```

##### Price distribution

```python
sns.boxplot(x=df_housing["price"])
```

```python
# Plotting the histogram
### Params
price_min = 100000
price_max = df_housing['price'].max()
bin_width = 100000
bins = range(int(price_min), int(price_max) + int(bin_width), int(bin_width))

### Plot
plt.figure(figsize=(14, 8))
sns.histplot(df_housing['price'], bins=bins, kde=False)

plt.title('Distribution of Property Prices')
plt.xlabel('Price [PLN]')
plt.ylabel('Frequency')

### Update labels to be in 100k PLN
labels = [f'{(label/1000)}k' for label in bins]
plt.xticks(bins, labels, rotation=45)

plt.show()
```

##### Price vs Total Area

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_area', y='price', data=df_housing)
plt.title('Price vs Total Area')
plt.xlabel('Total Area (mÂ²)')
plt.ylabel('Price')
plt.show()
```

##### Price vs Building Year

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='building_year', y='price', data=df_housing)
plt.title('Price vs Building Year')
plt.xlabel('Building Year')
plt.ylabel('Price')

plt.xlim(df_housing['building_year'].min(), df_housing['building_year'].max())
plt.ylim(100000, df_housing['price'].max())
plt.yticks(range(100000, int(df_housing['price'].max()) + 100000, 100000))

plt.show()

```

##### Price by District

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='district_norm', y='price', data=df_housing)
plt.title('Price by District')
plt.xlabel('District')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

##### Price by building type

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='building_type', y='price', data=df_housing)
plt.title('Price by Building Type')
plt.xlabel('Building Type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

##### Validate how do the categorical attrbiutes contribute to price

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='heating_type', y='price', data=df_housing)
plt.title('Price by Heating Type')
plt.xlabel('Heating Type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```
